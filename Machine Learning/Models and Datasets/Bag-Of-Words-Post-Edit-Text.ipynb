```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import io
import nltk
import re, string, timeit
from textblob import TextBlob
import csv
import numpy as np
```

## BoW for the training dataset of the post edit text

```python
stop_words = set(stopwords.words('english'))

inputFile = open("Database\ml-training-post-edit-text.txt", 'r', encoding='utf-8')
outputFile = open('Database\output-training.txt','w', encoding='utf-8')

lines = inputFile.readlines()
# print(lines)
inputFile.close()

stemming = nltk.stem.SnowballStemmer('english') # Stemming

questionIds = []

for line in lines:
    questionIds.append(line.partition(' ')[0]) # get the question ID
    print(line.partition(' ')[0])
    line = line.partition(' ')[2]
    line = line.translate(str.maketrans('', '', string.punctuation))
    words = line.split()

    lineAfterProcessing = ""

    for r in words:
        r = r.strip()
        if not r in stop_words:
            if not r[0].isdigit():
                if not r.startswith("http"):
                    lineAfterProcessing += " "+stemming.stem(r)

    outputFile.write(lineAfterProcessing+"\n")
#     print(lineAfterProcessing)

outputFile.close()
```

```python
with open('Database\output-training.txt', 'r',  encoding='utf-8') as corpusFile:
    corpus = corpusFile.readlines()

vectorizer = CountVectorizer()

vectorizer.fit(corpus)
# print(sorted(vectorizer.vocabulary_))

#print(corpus)
X = vectorizer.fit_transform(corpus)

vector = X.toarray()
# print(vector)
```

```python
arrayRow = len(vector)
arrayColumn = len(vector[0])

csvFile = open('Database\ml-training-post-text-vector.csv', 'w', newline='',  encoding='utf-8')
csvWriter = csv.writer(csvFile)

vocabulary = sorted(vectorizer.vocabulary_)
vocabulary.insert(0,'questionid')

csvWriter.writerow(vocabulary)

# csvWriter.writerows(vector)

# print(arrayRow)
# print(arrayColumn)

for row in range(arrayRow):

    tempquestionIds = questionIds[row]
#     print(tempquestionIds)

    tempVector = list(vector[row])

    tempVector.insert(0, tempquestionIds)

#     tempVector = np.asarray(tempVector)

#     print(tempVector)

    csvWriter.writerow(tempVector)
#     for column in range(arrayColumn):
#         csvWriter.writerow(str(vector[row][column]))

csvFile.close()
```
